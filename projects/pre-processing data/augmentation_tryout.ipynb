```python
from torchvision.transforms import Compose, Resize, GaussianBlur, ToTensor, RandomRotation, RandomGrayscale, Normalize, Grayscale, RandomCrop
from PIL import Image
import os

directory ='/datadrive/animals_training_dataset/crops/coyote/'
files_inside_s = os.listdir(directory)
#random.shuffle(files_inside_s)
files_inside_s = files_inside_s[0:10]
print(files_inside_s)
```

```python
img = Image.open('/datadrive/animals_training_dataset/crops/coyote/TUW09.03140219.JPG').convert("RGB")
img
```

```python
GaussianBlur(kernel_size=(231),sigma=(0.05, 2))(img)
```

```python
RandomRotation(degrees=(-45, 45))(img)
```

```python
Resize((img.height*2, img.width*2))(img)
```

```python
RandomGrayscale(20)(img)
```

```python
RandomCrop(size=(img.height,50))(img)
```

```python
ToTensor()(img)
```

```python
Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))(ToTensor()(img))
```

```python
import torch
torch.mean(ToTensor()(img), dim = (0, 1, 2))
torch.std(ToTensor()(img), dim = (0, 1, 2))
Normalize(torch.mean(ToTensor()(img), dim = (0, 1, 2)), torch.std(ToTensor()(img), dim = (0, 1, 2)))(ToTensor()(img))
```
